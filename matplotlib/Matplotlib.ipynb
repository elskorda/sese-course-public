# Multiple series in a single plot

```python
x = np.linspace(0.0, 2*np.pi, 50)
y1 = np.sin(x)
y2 = np.cos(x)
```

Multiple series can be added by adding arguments to the plot function:

```python
plt.plot(x, y1, 'r-', x, y2, 'b-')
plt.show()
```

# Changing series attributes

```python
plt.plot(x, y1, 'r-', x, y2, 'b-', linewidth=4.0) # Changing linewidth for all series
plt.show()
```

```python
line1, line2 = plt.plot(x, y1, 'r-', x, y2, 'b-')
line1.set_linewidth(4.0)
line2.set_linewidth(8.0)
plt.show()
```

```python
lines = plt.plot(x, y1, 'r-', x, y2, 'b-') # lines is a list of Line2D objects
plt.setp(lines, linewidth = 8.0) # Sets attributes for all lines 
plt.show()
```

# Multiple figures

Multiple figures (windows/views) can be created using the figure() function. Figures can be given numbers to be able to switch plotting from one figure/window to another.

```python
plt.figure(1)
plt.plot(x, y1, 'r-')

plt.figure(2)
plt.plot(x, y2, 'b-')

plt.show()
```

# Subplots

Often plots needs to be combined in a single plot. This is done with the subplots() function. Arguments to the the functions are given by a special integer consisting of 3 numbers. The first numbers gives the rows and columns for the plot. The last number specifies the id of the subplot. 

211 - [2x1] id = 1

212 - [2x1] id = 2

```python
plt.figure(1)
plt.subplot(211) # Subplot with 2 rows, 1 column and first plot.
plt.plot(x, y1, 'r-')

plt.subplot(212)
plt.plot(x, y2, 'b-') # Subplot with 2 rows, 1 column and second plot.

plt.show()
```

```python
plt.figure(1)
plt.subplot(121) # Subplot with 1 row, 2 columns and first plot.
plt.plot(x, y1, 'r-')

plt.subplot(122)
plt.plot(x, y2, 'b-') # Subplot with 1 row, 2 columns and second plot.

plt.show()
```

# Decorating plots

There are several functions for decorating plots in matplotlib. All functions in matlplotlib work on the currently active figure. Switching between figures/subplots is done by using the figure() and subplot() functions.

```python
plt.figure(1)
plt.subplot(211) # Switching to figure 211
plt.xlabel("x")
plt.ylabel("y1")
plt.plot(x, y1, 'r-')

plt.subplot(212) # Switching to figure 212
plt.xlabel("x")
plt.ylabel("y2")
plt.plot(x, y2, 'b-')

plt.show()
```

Text can be added to a plot using the text() function. Text are placed using coordinates of the plot. LaTEX commands can be used to display mathematical functions in the plot.

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.text(np.pi/2, 0.8, r'$sin(x=\frac{pi}{2})$', horizontalalignment='center')
plt.plot(x, y, 'r-')
plt.show()
```

Alignment can be changed using the horizontalalignment and verticalalignment arguments to the text() function.

Annotations can also be added to a plot using the annotate() function.

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.annotate('max', xy=(np.pi/2, 1), xytext=(np.pi/2, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05), horizontalalignment='center'
            )
plt.plot(x, y, 'r-')

plt.show()
```

Axis limits can be set by using the xlim() and ylim() functions:

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0.0, np.pi)
plt.ylim(0.0, 1.0)
plt.annotate('max', xy=(np.pi/2, 1), xytext=(np.pi/2, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(x, y, 'r-')
plt.savefig("test.eps")
plt.show()
%ls
```

Axis scales can also be changed from linear to log-scales using the xscale() and yscale() functions:

```python
x = np.arange(0,2.0, 0.01)
y = np.exp(x)

plt.figure(1)
plt.yscale('log')
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y, 'r-')

plt.show()
```

# More plot types

## Histogram - hist()

```python
np.random.seed(0)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

plt.hist(x, num_bins, normed=1)

plt.xlabel('Smarts')
plt.ylabel('Probability density')

plt.show()
```

## Bar diagrams

```python
np.random.seed(0)

# example data
x = np.arange(20)+1.0
y = np.random.rand(20)

plt.bar(x, y)
plt.xlim(0,21)
plt.xticks(range(1,20+1))

plt.show()
```

## Scatter plots

```python
for color in ['red', 'green', 'blue']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    plt.scatter(x, y, c=color, s=scale, label=color,
                alpha=0.3, edgecolors='none')

plt.legend()   # Produces a legend in the diageram.
plt.grid(True) # Adds a grid in the diagram

plt.show()
```

## Contour plots

```python
N = 100

x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)

z = np.sin(0.5*X*Y)

plt.contourf(X, Y, z, cmap="gray") # cmap defines the colormap used.
plt.colorbar() # Adds a colorbar to the plot

plt.show()
```

## Displaying images

```python
import matplotlib.cbook as cbook

image_file = cbook.get_sample_data('ada.png')
image = plt.imread(image_file)

plt.imshow(image)
plt.axis('on')  # clear x- and y-axes
plt.show()
```

Interpolation of images (smoothing) can also be controlled:

```python
# from the docs:

# If interpolation is None, default to rc image.interpolation. See also
# the filternorm and filterrad parameters. If interpolation is 'none', then
# no interpolation is performed on the Agg, ps and pdf backends. Other
# backends will fall back to 'nearest'.
#
# http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow

methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

np.random.seed(0)
grid = np.random.rand(4, 4)

fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(grid, interpolation=interp_method, cmap='viridis')
    ax.set_title(interp_method)

plt.show()
```

## Stream plots

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

fig0, ax0 = plt.subplots()
strm = ax0.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
fig0.colorbar(strm.lines)

fig1, (ax1, ax2) = plt.subplots(ncols=2)
ax1.streamplot(X, Y, U, V, density=[0.5, 1])

lw = 5*speed / speed.max()
ax2.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)

fig2, (ax3, ax4) = plt.subplots(ncols=2)
ax3.contourf(X, Y, U, cmap=plt.cm.RdBu)
ax4.contourf(X, Y, V, cmap=plt.cm.RdBu)

plt.show()
```

```python
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```

## 3D surface plot

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

N = 100

x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)

z = np.sin(0.5*X*Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, z, cmap=plt.cm.RdBu)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_zlim(-5, 5)
plt.show()
```
